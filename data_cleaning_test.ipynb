```python
# Calculating total no. of null values in each column as we saw there multiple nan values in columns
df.isnull().sum()
```

```python
#Dropping the columns which is not usable or have high percentage of null values in order to reduce the size of data 
# which will decrease the processing time as well
df.drop(columns=['url','region_url','VIN','image_url','description','county','posting_date','size','lat','long'],inplace=True)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
# Number of rows having more than one null value in it i.e. 2 or more columns having null value in same row
df.isnull().sum(axis=1).gt(1).sum()
```

```python
df.isnull().sum(axis=1).gt(2).sum()
```

```python
df.isnull().sum(axis=1).gt(3).sum()
```

```python
df.isnull().sum(axis=1).gt(4).sum()
```

```python
#filtering the data to keep only rows which has 2 or less number of null values in it
df = df[df.isnull().sum(axis=1) <= 2]
```

```python
df.reset_index(drop=True, inplace=True)
```
